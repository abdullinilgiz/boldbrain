```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_caps_embed_path = '../data/caps_embeds/train_caps_embeds_sub1.npy'
train_caps_embed_neg_path = '../data/caps_embeds/train_caps_embeds_negative_sub1.npy'

train_caps_embed = np.load(train_caps_embed_path)
train_caps_embed_neg = np.load(train_caps_embed_neg_path)
```

```python
test_caps_embed_path = '../data/caps_embeds/test_caps_embeds_sub1.npy'
test_caps_embed_neg_path = '../data/caps_embeds/test_caps_embeds_negative_sub1.npy'

test_caps_embed = np.load(test_caps_embed_path)
test_caps_embed_neg = np.load(test_caps_embed_neg_path)
```

```python
train_caps_embed.shape
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

variances = np.var(train_caps_embed, axis=0)[0]
x = np.arange(len(variances))
y = variances
y.sort()

ax1.plot(x, y)
ax1.set_xlabel('Индекс измерения')
ax1.set_ylabel('Дисперсия')
ax1.set_title('train_caps_embed')
ax1.set_ylim(0, 5)

variances = np.var(train_caps_embed_neg, axis=0)[0]
x = np.arange(len(variances))
y = variances
y.sort()

ax2.plot(x, y)
ax2.set_xlabel('Индекс измерения')
ax2.set_ylabel('Дисперсия')
ax2.set_title('train_caps_embed_neg')
ax2.set_ylim(0, 5)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

variances = np.var(test_caps_embed, axis=0)[0]
x = np.arange(len(variances))
y = variances
y.sort()

ax1.plot(x, y)
ax1.set_xlabel('Индекс измерения')
ax1.set_ylabel('Дисперсия')
ax1.set_title('test_caps_embed')
ax1.set_ylim(0, 5)

variances = np.var(test_caps_embed_neg, axis=0)[0]
x = np.arange(len(variances))
y = variances
y.sort()

ax2.plot(x, y)
ax2.set_xlabel('Индекс измерения')
ax2.set_ylabel('Дисперсия')
ax2.set_title('test_caps_embed_neg')
ax2.set_ylim(0, 5)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

mean_embed = np.mean(train_caps_embed, axis=0)[0]
x = np.arange(len(mean_embed))
y = mean_embed
y.sort()

ax1.plot(x, y)
ax1.set_xlabel('Индекс измерения')
ax1.set_ylabel('Mean')
ax1.set_title('train_caps_embed')
ax1.set_ylim(-6, 6)

mean_embed = np.mean(train_caps_embed_neg, axis=0)[0]
x = np.arange(len(mean_embed))
y = mean_embed
y.sort()

ax2.plot(x, y)
ax2.set_xlabel('Индекс измерения')
ax2.set_ylabel('Mean')
ax2.set_title('train_caps_embed_neg')
ax2.set_ylim(-6, 6)

plt.show()
```
